```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def cross_entropy(p):
    return -np.log(p)
ps = np.linspace(0.005, 1, 100)
ces = [cross_entropy(p) for p in ps]

plt.plot(ps, ces)
plt.title('cross-entropy loss v.s. probability of ground truth class')
plt.ylabel('loss')
plt.xlabel('probability')
plt.grid()
```

```python
ps = list(np.linspace(0.1, 0.9, 9))+list(np.linspace(0.91, 1, 10))
ces = [cross_entropy(p) for p in ps]
df = pd.DataFrame([ces, list(ps)]).T
df.columns = ['loss', 'p']
df['loss'] = df.loss.round(2)
df
```

```python
df.to_csv('cross_entropy_vs_p.csv', index=False)
```
